Abaixo segue um exemplo de um código funcional em *Python* rodando em um *notebook*:

```python
n1 = 15
n2 = 5
n3 = n1 + n2

print(f'Qual o nome do filme?\n Resposta = O código dá {n3}.')
```

Você pode limpar a saida da célula (resultado) para editar o codigo para fazer outro codigo, clique nos 3 pontinhos (...) ao lado do resultado do codigo e selecione "Limpar saida das celulas" ou pode copiar a saida clicando em "Copiar saida da celula" para usar o resultado em algum lugar

O numerozinho ao lado do tempo de execucao ex: [6], é indicando a ordem de execucao das celulas, pois a ordem de execucao das celulas muda o resultado

Repare que o código-fonte das células consideram o resultado das células anteriores do mesmo notebook. Observe no código a seguir:
